## Youtube Spam Detection using RNN

## Loading necessary Libraries

```python
import numpy as np
import pandas as pd
from tensorflow.keras import Input,Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense,Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
df = pd.read_csv("merged_youtube_comments.csv")
```

```python
df.sample(5)
```

```python
df["CONTENT"] = df["CONTENT"].str.lower()
df["CONTENT"] = df["CONTENT"].replace(r'[^a-z0-9\s]', '', regex=True)
```

```python
df.sample(1)
```

### Text Preprocessing

```python
max_words = 1000
max_len = 100

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(df["CONTENT"])


sequences = tokenizer.texts_to_sequences(df["CONTENT"])
x = pad_sequences(sequences,maxlen=max_len)
y = df["CLASS"]
```

### Train - Test Split

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

### Model Building

```python
vocab_size = min(len(tokenizer.word_index) + 1, max_words)
inputs = Input(shape=(max_len,))
x = Embedding(vocab_size,32)(inputs)
x = SimpleRNN(64,return_sequences=True)(x)
x = Dropout(0.1)(x)
x = SimpleRNN(32)(x)
x = Dropout(0.1)(x)
x = Dense(64,activation="relu")(x)
x = Dropout(0.2)(x)
output = Dense(1,activation="sigmoid")(x)

model = Model(inputs=inputs,outputs=output)
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

### Training Model 

```python
early = EarlyStopping(monitor="loss",patience = 3,restore_best_weights = True)
```

```python
hist = model.fit(
    xtrain,
    ytrain,
    epochs = 10,
    callbacks = [early],
    validation_split=0.2
)
```

### Visualization

```python
plt.plot(hist.history["val_loss"],label="Training Loss")
plt.plot(hist.history["loss"],label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Epoch vs Loss")
plt.legend()
plt.show()
```

```python
plt.plot(hist.history["val_accuracy"],label="Training Accuracy")
plt.plot(hist.history["accuracy"],label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Epoch vs Accuracy")
plt.legend()
plt.show()
```

### Confusion Matrix

```python
y_pred_probs = model.predict(xtest)
y_pred = np.argmax(y_pred_probs, axis=1)

# Confusion matrix
cm = confusion_matrix(ytest, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix - Youtube Spam Detection")
plt.show()
```
