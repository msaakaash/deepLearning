```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

### Sigmoid Activation Function

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

```python
def plot_sigmoid():
    x = np.linspace(-10,10,100)
    y = 1 / (1+np.exp(-x))
    plt.plot(x,y)
    plt.xlabel('Input')
    plt.ylabel('Sigmoid Output')
    plt.title('Sigmoid Activation Function')
    plt.grid(True)
    plt.show()
plot_sigmoid()
```

### Tanh

```python
def tanh(x):
    return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
tanh(-56)
```

```python
tanh(1)
```

```python
def plot_tanh():
    x = np.linspace(-10,10,100)
    y = tanh(x)
    plt.plot(x,y)
    plt.title("Hyperbolic Tangent")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.show()
plot_tanh()
```

### Relu 

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def plot_relu():
    x = np.linspace(-10,10,100)
    y = relu(x)
    plt.plot(x,y)
    plt.title("Hyperbolic Tangent")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.show()
plot_relu()
```

### Leaky Relu

```python
def leaky_relu(x,alpha=0.1):
    return np.maximum(x,alpha*x)
```

```python
def plot_leakyrelu():
    x = np.linspace(-10,10,100)
    y=leaky_relu(x)
    plt.plot(x,y)
    plt.title("Leaky Relu")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.show()
plot_leakyrelu()
```
