## Applicances Energy Prediction

### Loading the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import plot_model
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
df = pd.read_csv("energydata_complete.csv")
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
df["date"] = pd.to_datetime(df["date"]).astype('int64')
```

```python
df.sample(3)
```

### Scaling

```python
x = df.drop("Appliances",axis=1)
y = df["Appliances"]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
xscaled = scaler.fit_transform(x)
yscaled = scaler.fit_transform(df[["Appliances"]])
train_size = int(len(xscaled)*0.8)
```

### Creating Sequences

```python
def create_sequences(x,y,seq_len):
    xseq , yseq = [] , []
    for i in range(len(x)-seq_len):
        xseq.append(x[i:i+seq_len])
        yseq.append(y[i+seq_len])
    return np.array(xseq),np.array(yseq)

sequence_length = 10
```

```python
xtrain_data = xscaled[:train_size]
ytrain_data = yscaled[:train_size]

xtest_data = xscaled[train_size-sequence_length:]
ytest_data = yscaled[train_size-sequence_length:]

xtrain,ytrain = create_sequences(xtrain_data,ytrain_data,sequence_length)
xtest,ytest = create_sequences(xtest_data,ytest_data,sequence_length)
```

### Model Building

```python
model = Sequential([
    LSTM(50,return_sequences=True,input_shape=(sequence_length,28),recurrent_dropout=0.1),
    Dropout(0.1),
    LSTM(25,recurrent_dropout=0.1),
    Dropout(0.1),
    Dense(12),
    Dense(1)
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss = "mean_squared_error")
```

```python
earlyStopping = EarlyStopping(
    monitor = "val_loss",
    patience = 10,
    restore_best_weights = True
)
```

```python
lr_scheduler = ReduceLROnPlateau(
    monitor = "val_loss",
    patience = 5,
    factor = 0.5,
    min_lr=0.0001
)
```

### Training the model

```python
hist = model.fit(
    xtrain,
    ytrain,
    epochs = 15,
    callbacks = [earlyStopping,lr_scheduler],
    validation_split=0.2
)
```

```python
## Visualization
```

```python
plt.plot(hist.history["val_loss"],label="Training Loss")
plt.plot(hist.history["loss"],label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Epoch vs Loss")
plt.legend()
plt.show()
```

```python
training_predictions = model.predict(xtrain)
testing_predictions = model.predict(xtest)
```

```python
training_predictions = scaler.inverse_transform(training_predictions)
y_train_inverse = scaler.inverse_transform(ytrain)
testing_predictions = scaler.inverse_transform(testing_predictions)
y_tesn_inverse = scaler.inverse_transform(ytest)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(y_tesn_inverse, label='Actual Demand', color='blue', 
alpha=0.7)
plt.plot(testing_predictions, label='Predicted Demand', 
color='red', alpha=0.7)
plt.title('Actual vs. Predicted Demand - Test Data')
plt.ylabel('Number of Orders')
plt.xlabel('Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('test_predictions.png')
plt.show()
```
