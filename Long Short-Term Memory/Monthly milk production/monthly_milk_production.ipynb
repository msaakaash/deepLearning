## Loading necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout
```

## Data Preparation and Scaling

```python
data = pd.read_csv("monthly_milk_production.csv")
data.head(5)
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data.set_index('Date',inplace=True)
production = data["Production"].astype(float).values.reshape(-1,1)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(production)
scaled_data
```

## Creating Sequences

```python
window_size = 12
x = []
y = []

target_dates = data.index[window_size:]

for i in range(window_size,len(scaled_data)):
    x.append(scaled_data[i-window_size:i,0])
    y.append(scaled_data[i,0])

x = np.array(x)
y = np.array(y)
```

## Train - Test - Split

```python
xtrain,xtest,ytrain,ytest,dates_train,dates_test = train_test_split(x,y,target_dates,test_size=0.2,shuffle=False)
xtrain = xtrain.reshape((xtrain.shape[0],xtrain.shape[1],1))
xtest = xtest.reshape((xtest.shape[0],xtest.shape[1],1))
```

## Building LSTM model

```python
model = Sequential()
model.add(LSTM(units=128,return_sequences=True,input_shape=(xtrain.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=128))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer = 'adam',loss='mean_squared_error')
```

## Training the model

```python
hist = model.fit(xtrain,ytrain,epochs = 100,batch_size=32,validation_split=0.1)

predictions = model.predict(xtest)
predictions = scaler.inverse_transform(predictions).flatten()
ytest = scaler.inverse_transform(ytest.reshape(-1,1)).flatten()

rmse = np.sqrt(np.mean((ytest - predictions)**2))
print(f'RMSE: {rmse:.2F}')
```

## Epoch vs Loss

```python
# Extract loss values from the history object
train_loss = hist.history['loss']  # Training loss
val_loss = hist.history['val_loss']  # Validation loss
epochs = range(1, len(train_loss) + 1)  # Number of epochs

# Plot the loss graph
plt.figure(figsize=(8, 5))
plt.plot(epochs, train_loss, label="Training Loss", marker="o", linestyle="-")
plt.plot(epochs, val_loss, label="Validation Loss", marker="o", linestyle="--")

# Labels and Title
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Epoch vs Loss")
plt.legend()
plt.grid(True)
plt.show()
```
