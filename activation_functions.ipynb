```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sigmoid

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def derivative_sigmoid(y):
  return sigmoid(y)*(1-sigmoid(y))
```

```python
sigmoid(1)
```

```python
def plot_sigmoid():
  x = np.linspace(-10,10,100)
  y = sigmoid(x)
  plt.plot(x,y)
  plt.title("Sigmoid Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()

  z = derivative_sigmoid(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_sigmoid()
```

### Tanh

```python
def tanh(x):
  return ((np.exp(x)-np.exp(-x)))/(np.exp(x)+np.exp(-x))
```

```python
tanh(-56)
```

```python
def derivative_tanh(x):
  return 1-tanh(x)**2
```

```python
def plot_tanh():
  x = np.linspace(-10,10,100)
  y = tanh(x)
  plt.plot(x,y)
  plt.title("Tanh Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()

  z =  derivative_tanh(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_tanh()
```

### Relu

```python
def relu(x):
  return np.maximum(0,x)
```

```python
def derivative_relu(x):
  return np.where(x > 0, 1, 0)
```

```python
def plot_relu():
  x = np.linspace(-10,10,100)
  y = relu(x)
  plt.plot(x,y)
  plt.title("Relu Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()

  z= derivative_relu(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_relu()
```

### Leaky Relu

```python
def lrelu(x,alp=0.1):
  return np.maximum(alp*x,x)
```

```python
def derivative_lrelu(x,alp=0.1):
  return np.where(x > 0,1, alp)
```

```python
def plot_lrelu():
  x = np.linspace(-10,10,100)
  y = lrelu(x)
  plt.plot(x,y)
  plt.title("Relu Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()


  z = derivative_lrelu(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_lrelu()
```

### ELU

```python
def elu(x,alpha=1):
  return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
def derivative_relu(x,alpha=1):
  return np.where(x > 0, 1, alpha*elu(x))
```

```python
def plot_elu():
  x = np.linspace(-10,10,100)
  y = elu(x)
  plt.plot(x,y)
  plt.title("Relu Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()

  z = derivative_relu(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_elu()
```

### Soft Max

```python
def softmax(x):
  x_shifted = x - np.max(x)
  exp_x = np.exp(x_shifted)
  return exp_x / np.sum(exp_x)
```

```python
def plot_sof():
  x = np.linspace(-10,10,100)
  y = elu(x)
  plt.plot(x,y)
  plt.title("SoftMax Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_sof()
```

```python
def softplus(x):
  return np.log(1+np.exp(x))
```

```python
def derivative_softplus(x):
  return 1/(1+np.exp(-x))
```

```python
def plot_sof():
  x = np.linspace(-10,10,100)
  y = softmax(x)
  plt.plot(x,y)
  plt.title("Softplus Function")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()


  z=derivative_softplus(x)
  plt.plot(x,z)
  plt.title("Derivative")
  plt.xlabel("Input")
  plt.ylabel("Output")
  plt.show()
plot_sof()
```
